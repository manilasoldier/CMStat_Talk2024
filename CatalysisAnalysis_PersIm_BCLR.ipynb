```python
import detectda as dtda
import bclr
import pickle
import matplotlib.pyplot as plt
from skimage.exposure import equalize_hist
from skimage import filters
import numpy as np
import ruptures as rpt
```

```python
file = open("030303_P2_Noisy.pkl", 'rb')
vidpol = pickle.load(file)
```

```python
video, polygon, bounds = vidpol.values()
```

```python
plt.imshow(equalize_hist(video[25]), cmap='gray')
```

```python
dtda_obj = dtda.ImageSeriesPlus(video, polygon, div=32, n_jobs=4)
dtda_obj.fit(sigma=2)
```

```python
dtda_obj.plot_im(30, thr=0.1)
```

```python
ps = dtda_obj.get_pers_stats()
```

```python
np.argsort(ps['pers_entr_0_life'])
```

```python
dtda_obj.get_lifetimes()
dtda_obj.plot_im(40, thr=0.1)
```

```python
dtda_obj.get_pers_im(6, 6 , 0 , 2)
prior_cov = np.diag(np.repeat(0.2, 36))
```

```python
rng = np.random.default_rng(883481)
bclr_dtda = bclr.MultiBayesCC(dtda_obj.pis, cps=8, prior_cov=prior_cov, n_iter=1000, rng=rng)
cp_df = bclr_dtda.fit_predict()
```

```python
cp_df
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'Nunito'

for i in range(5):
    if i == 0:
        x = np.arange(1, cp_df['Location'][i+1]+1)
        y = dtda_obj.pis @ np.expand_dims(bclr_dtda.bccs_[i].post_beta_mean, axis=1)
        y = y[:int(cp_df['Location'][i+1]), 0]
    elif i == 4:
        x = np.arange(cp_df['Location'][i-1]+2, len(video)+1)
        y = dtda_obj.pis @ np.expand_dims(bclr_dtda.bccs_[i].post_beta_mean, axis=1)
        y = y[int(cp_df['Location'][i-1]+1):(len(video)), 0]
    else:
        x = np.arange(cp_df['Location'][i-1]+2, cp_df['Location'][i+1]+1)
        y = dtda_obj.pis @ np.expand_dims(bclr_dtda.bccs_[i].post_beta_mean, axis=1)
        y = y[int(cp_df['Location'][i-1]+1):int(cp_df['Location'][i+1]), 0]

    plt.figure().set_figwidth(10)
    plt.plot(x, y, color='#1E88E5')
    plt.xlabel('Frame $i$')
    plt.ylabel(r'$\mathbf{x}_{\kappa}^{\top}{\bar{\beta}_'+str(i+1)+'}$')
    plt.title('Posterior mean and change estimate '+str(i+1))
    val = cp_df['Location'][i]
    plt.axvline(val+1, color='black', linewidth=1.7)
    x1, x2 = np.quantile(bclr_dtda.bccs_[i].post_k, (0.025, 0.975))
    plt.fill_betweenx(np.linspace(min(y)-0.02, max(y)+0.02, 100), x1, x2, color='red', alpha=0.5)
    plt.ylim(min(y)-0.02, max(y)+0.02)
    plt.savefig("Nanoparticle_change "+str(i+1)+".png")
```

```python
cp1 = np.arange(0, int(cp_df['Location'][1]))
cp4 = np.arange(int(cp_df['Location'][2]), int(cp_df['Location'][4]))
```

```python
dtda_obj.proc_pers_im(bclr_dtda.bccs_[0].post_beta, quantiles=(0.3, 0.7), indices=cp1)
```

```python
import imageio
l_names = []
for i in cp1:
    print(i)
    dtda_obj.plot_pi_sig(i, betas_feat='both')
    name_i = "First_Change/PI_cp1_030303_P2_Im"+str(i+1)+".png"
    l_names += [name_i]
    if i <= int(cp_df['Location'][0]):
        plt.annotate(str("Pre-change"), (0.35*259, 0.04*348), color='white', size=12, family="Nunito", weight="heavy")
    else: 
        plt.annotate(str("Post-change"), (0.32*259, 0.04*348), color='white', size=12, family="Nunito", weight="heavy")
    plt.savefig(name_i)
    plt.close()

with imageio.get_writer('First_Change/PI_cp1_030303_P2.gif', mode='I') as writer:
    for filename in l_names:
        image = imageio.imread(filename)
        writer.append_data(image)
```
